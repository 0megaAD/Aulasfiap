```python
N1, N2, N3, N4 = input("Insira a Nota: ").split()
N1 = float(N1)
N2 = float(N2)
N3 = float(N3)
N4 = float(N4)

def nota(N1, N2, N3, N4):
  media = ((N1 * 2) + (N2 * 3) + (N3 * 4) + (N4 * 1)) / 10
  print(f'Média: {media}')

  if media >= 7:
    print('Aprovado')

  elif media >= 5:
    print('Aluno em exame')
    exame = float(input('Insira a nota do exame: '))
    media = (media + exame) / 2
    print(f'Nota do exame: {exame}')

    if media >= 5:
      print('Aprovado')

    else:
      print('Reprovado')

  else:
    print('Reprovado')

  print(f'\n\nMédia Final: {media}')

nota(N1, N2, N3, N4)
```
