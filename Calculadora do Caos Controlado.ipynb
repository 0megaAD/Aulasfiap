```python
def media(A,B,C):
  PA = 2
  PB = 3
  PC = 5
  media = ((PA*A + PB*B + PC*C)/(PA + PB + PC))
  print(f"A Media é {media}")

def desvio_P(valores):
    return variancia(valores) ** 0.5

def variancia(valores):
  media = sum(valores) / len(valores)
  variancia = sum((x - media) ** 2 for x in valores)
  return variancia
```

```python
A = float(input())
B = float(input())
C = float(input())
media(A, B, C)
```

```python
numeros = []

for i in range(4):
    n = int(input("Insira os números da distribuição: "))
    numeros.append(n)

print("Variância:", variancia(numeros))
print("Desvio padrão:", desvio_P(numeros))
```

```python
notas = []
desvios = []
variancia = []

for i in range(4):
    n = int(input("Insira os números da distribuição: "))
    notas.append(n)

media = sum(notas) / len(notas)

for n in notas:
    desvios.append(n - media)

for d in desvios:
    variancia.append(d ** 2)

print("Notas:", notas)
print("Média:", media)
print("Desvios:", desvios)
print("Variância:", sum(variancia) / len(variancia))
```
